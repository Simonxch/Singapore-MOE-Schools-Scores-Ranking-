```python
# task 8B-1: Drop missing values with dropna

mydf = pd.read_excel('data/singstats_maritalstatus.xlsx', index_col='Variables', na_values='-')

print(f'**** First 10 rows of original dataset ****')
print(mydf.iloc[:10,:7]);  print()

mydf1 = mydf.isnull()
mydf2 = mydf.dropna(axis=1)
print(f'**** Remaining dataset after dropping columns with missing data ****')
print(mydf2.iloc[:10,:6]);
```

```python
# task 8C-1: Drop missing values with dropna
mydf = pd.DataFrame(pd.read_csv('data/hdbresaleindex_missing.csv', index_col='quarter'))

print(f'***** First 10 rows of original dataswet *****')
print(mydf.iloc[:10]);  print()

print(f'The sum of the null cells count: {sum(mydf.isnull().any(axis=1))}')
print()

df_missing = mydf[mydf.isnull().any(axis=1)]
print(f'**** Rows with missing values -- df_missing ****')
print(df_missing);  print()


df1 = mydf.fillna(50.0)
print(f'***** Dataset after scalar fill operation *****')
print(df1.iloc[:12]);  print()

df2 = mydf.fillna(method='pad')
print(f'***** Dataset after forward fill operation *****')
print(df2.iloc[:12]);  print()

df3 = mydf.fillna(method='bfill')
print(f'***** Dataset after backward fill operation *****')
print(df3.iloc[:12]);  print()

df4 = mydf.fillna(mydf.mean(axis=0))
print(f'***** Dataset after filling with means *****')
print(df4.iloc[:12]);  print()
```
