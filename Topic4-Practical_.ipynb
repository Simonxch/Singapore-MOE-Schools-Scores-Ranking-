# Task 4.	Working with text

```python
# task 4-1
import numpy as np
x = np.arange(0.0, 5.0, 0.01)
y = np.cos(2*np.pi*x)

plt.plot(x,y, 'r')
plt.xlim(0,6)
plt.ylim(-1.5,1.5)
plt.text(2, 1, 'x=2, y=1')
plt.text(5, 1, 'x=5, y=1', bbox={'facecolor':'red', 'alpha':0.5, 'pad':5})
plt.show()
```

```python
# task 4-3
import numpy as np
x = np.linspace(0, 2*np.pi, 128)
y1 = np.cos(x) # y values for c1
y2 = np.sin(x) # y values for c2

fig = plt.figure(figsize=(3, 3))
fig.suptitle('Matplotlib Lab', fontsize=16, ha='center')

ax = fig.add_subplot(111)
ax.plot(x, y1, label = "c1",color='#2E5984')
ax.plot(x, y2, label = "c2", color='green')

fig.tight_layout(pad=1.5) # adjust the space in between

ax.set_title('Working with text demo')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.legend()
plt.show()
```

```python
# reference for self-learning
x = np.linspace(0.0, 5.0, 501)

fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, sharey=True)
ax1.plot(x, np.cos(6*x) * np.exp(-x))
ax1.set_title('damped')
ax1.set_xlabel('time (s)')
ax1.set_ylabel('amplitude')

ax2.plot(x, np.cos(6*x))
ax2.set_xlabel('time (s)')
ax2.set_title('undamped')

fig.suptitle('Different types of oscillations', fontsize=16)
plt.show()
```

# 7.	Scatter Plots

```python
# task 7-1
temp = pd.read_csv('data/surface-air-temperature-monthly-mean.csv')
rainfall = pd.read_csv('data/rainfall-monthly-number-of-rain-days.csv')

df_merged = pd.merge(temp, rainfall, on='month', how="inner")
df_merged

# split the data into x and y arrays
x = df_merged['mean_temp']
y = df_merged['no_of_rainy_days']

plt.scatter(x,y,c='g')
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), 'r-')

plt.title('Monthly rainfall and temperatures since Jan 1982')
plt.xlabel('Mean Monthly temperatures')
plt.ylabel('Number of rainy days')
plt.show()
```
